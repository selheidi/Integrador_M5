3.21 The 4 C's of Data Cleaning: Correcting, Completing, Creating, and Converting.

1-  Correcting: non-acceptable data inputs( potential outliers)
2- Completing: There are null values or missing data
3- Creating:  use existing features (características) to create new features to determine if they provide new signals to predict our outcome
4- Converting: formatting (si hay fechas se traen como objetos y hay que cambiarlas a datetype)
